```python
from pytube import YouTube

url = "https://www.youtube.com/watch?v=qURcKTur49w"

yt = YouTube(url)

# 抓取標題試試看
yt.title
```

```python
yt
```

```python
# 下載純音檔-->可能會拒絶存取
streams = yt.streams
audio = streams.get_audio_only()
obj = audio.download(filename="test.mp4")
```

```python
# 載入辨識模型
import whisper

model = whisper.load_model("large")
```

```python
# 必需安裝 ffmpeg.exe (建議在Colab上執行或者依網站說明安裝)
# 對下載的音檔進行辨識
# https://github.com/openai/whisper/blob/main/README.md
# https://platform.openai.com/docs/overview
result = model.transcribe("test.mp4")  # , language='zh'
```
