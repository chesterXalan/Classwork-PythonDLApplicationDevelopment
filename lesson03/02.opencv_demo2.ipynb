```python
import cv2
import numpy as np
import matplotlib.pylab as plt

img = cv2.imread("imgs/Holloways_beach.jpg")
plt.imshow(img)
plt.show()
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # 

plt.imshow(img2, cmap="gray")
plt.show()
```

```python
img3 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img3)
plt.show()
```

```python
print("kernel-1")
kernel1 = np.array(([0, -1, 0], [-1, 5, -1], [0, -1, 0]), dtype="int")

img3 = cv2.filter2D(img2, -1, kernel1)
plt.imshow(img3)
```

```python
print("kernel-2")
kernel2 = np.array(
    ([0.0625, 0.0125, 0.0625], [0.125, 0.35, 0.125], [0.0625, 0.125, 0.0625]),
    np.float32,
)

img4 = cv2.filter2D(img2, -1, kernel2)
plt.imshow(img4)
```

```python
print(np.array(img2))
print("--------------------------")
print(np.array(img3))
print("--------------------------")
print(np.array(img4))
print("--------------------------")
```
