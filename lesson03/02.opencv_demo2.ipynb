```python
import cv2
import numpy as np
import matplotlib.pylab as plt

img = cv2.imread("imgs/Holloways_beach.jpg")
plt.imshow(img)
plt.show()
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # 轉為灰階

plt.imshow(img2, cmap="gray")
plt.show()
```

```python
img3 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img3)
plt.show()
```

```python
print("建立kernel-1、銳利化")
kernel1 = np.array(([0, -1, 0], [-1, 5, -1], [0, -1, 0]), dtype="int")

img3 = cv2.filter2D(img2, -1, kernel1)
plt.imshow(img3)
```

```python
print("建立kernel-2、模糊化")
kernel2 = np.array(
    ([0.0625, 0.0125, 0.0625], [0.125, 0.35, 0.125], [0.0625, 0.125, 0.0625]),
    np.float32,
)

img4 = cv2.filter2D(img2, -1, kernel2)
plt.imshow(img4)
```

```python
print(np.array(img2))
print("--------------------------")
print(np.array(img3))
print("--------------------------")
print(np.array(img4))
print("--------------------------")
```

```python
# 縮放圖片
"""
| 參數
cv2.INTER_AREA : 常用在縮小圖片
cv2.INTER_CUBIC : 常用在放大圖片、4x4像素鄰域的雙三次插值
cv2.INTER_NEAREST : 最近鄰插值
cv2.INTER_LINEAR : 雙線性插值(默認)
cv2.INTER_LANCZOS4 : 8x8像素鄰域的Lanczos插值 |
"""
# cv2.resize(img, (w, h))
img5 = cv2.resize(img, (100, 50), cv2.INTER_AREA)
plt.imshow(img5)
cv2.imwrite("smallImg.png", img5)
```
