```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1, 3, 3], [1, 4, 31], [1, 1, 1], [1, 0, 2]])
y = np.array([[1], [1], [-1], [-1]])
```

```python
w = (np.random.random([3, 1]) - 0.5) * 2  # 3*1個亂數
```

```python
print(np.dot(X, w))
print(np.sign(np.dot(X, w)))
```

```python
lr = 0.5  # 學習率learning rate:調整的幅度，建議小
out = 0  # y hat:預測值


def update():
    global X, y, w, lr
    out = np.sign(np.dot(X, w))  # W@X--> 轉換
    w_c = lr * (X.T.dot(y - out)) / int(X.shape[0])  # 要新增的W差距
    w = w + w_c
```

```python
# X = np.array([[1, 3, 3], [1, 4, 31], [1, 1, 1], [1, 0, 2]])

for i in range(20):
    update()
    x1 = [3, 4]
    y1 = [3, 3]
    x2 = [1, 0]
    y2 = [1, 2]

    k = -w[1] / w[2]
    d = -w[0] / w[2]
    print("k=", k)
    print("d=", d)

    xdata = (0, 5)
    plt.figure()
    plt.plot(xdata, xdata * k + d, "r")  # 為了標示更新后的權重
    plt.scatter(x1, y1, c="b")
    plt.scatter(x2, y2, c="y")
    plt.show()
```
